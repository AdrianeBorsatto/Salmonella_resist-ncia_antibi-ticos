# Testes Microbiológicos de salmonella quanto a resistência à antibióticos

```python
from pandas.core.indexes.base import Index
dados = pd.read_excel('/content/Antibiograma. Salmonelas filtro (1).xlsx', usecols="B:O", header=1)
dados
```

```python
dados= dados.dropna()
dados
```

```python
print('O nome das colunas são os antibióticos usado na pesquisa:', dados.columns)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Criando um dicionário, onde R =0 e S= 1, representam resistência e sensibilidade, respectivamente

```python
traducao_dic= {
    'R':0,
    'S':1
}
```

```python
dados_modificados=dados[['Isolado', 'TOBRAMICINA ', 'RIFAMPICINA ',
       'SULFAZOTRIM ', 'FLORFENICOL ', 'METRONIDAZOL',
       'ÀC. CLAVULONICO + AMOXILINA ', 'CEFALEXINA ', 'CEFOTAXIMA  ',
       'TETRACICLINA ', 'LINEZOLIDA ', 'CLINDAMICINA ', 'IMIPENEM ', 'ENROFLOXACINA ']].replace(traducao_dic)
dados_modificados.head()
```

```python
dados_final= dados_modificados.dropna()
```

```python
dados_final.info()
```

```python
matriz_corr=dados_final.corr()
matriz_corr
```
